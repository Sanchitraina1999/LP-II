```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
data = make_blobs( n_samples=200, n_features=2, centers = 4, cluster_std= 1.5, random_state = 101)
```

```python
a=[]
b=[]
for i in data[0]:
    a.append(i[0])
    b.append(i[1])
    
plt.scatter(a,b,c=data[1])
```

```python
data[0]
```

```python
data[1]
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 4)
kmeans.fit(data[0])
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
plt.scatter(a,b,c=data[1])
plt.scatter( 3.70096449,  6.97822388, s=200, marker='s')
plt.scatter(-9.46293455, -6.56227402, s=200, marker='s')
plt.scatter( 0.17488363,  1.94463959, s=200, marker='s')
plt.scatter(-3.94290979,  7.81946035, s=200, marker='s')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
clustering = DBSCAN(eps=1, min_samples=3).fit(data[0])
clustering.labels_
```

```python
fig , ( ax1,ax2,ax3 )= plt.subplots(1,3,sharey=True,figsize=(10,6))
#Sample data created using make_blobs
ax1.set_title('Sample Data')
ax1.scatter(a,b,c=data[1],cmap="rainbow")
ax2.set_title('KMEANS')
ax2.scatter(a,b,c=kmeans.labels_,cmap="rainbow")
ax3.set_title('DBSCAN')
ax3.scatter(a,b,c=clustering.labels_,cmap="rainbow")
```
