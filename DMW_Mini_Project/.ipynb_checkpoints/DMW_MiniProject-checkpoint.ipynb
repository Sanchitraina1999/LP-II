```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report
from sklearn import metrics
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('./heart_new.csv')
df.head(10) #print first 10 rows
```

```python
f=open("./heart-disease.names")
print(f.read())
```

```python
df.info() #concise summary of the dataFrame
```

```python
df.isnull().sum() #missing values in the dataset
```

```python
#DATA CLEANING
df['age'].fillna(df['age'].mean(), inplace=True)
df['trestbps'].fillna(df['trestbps'].mean(), inplace=True)
df['chol'].fillna(df['chol'].mean(), inplace=True)
df['thalach'].fillna(df['thalach'].mean(), inplace=True)
df['oldpeak'].fillna(df['oldpeak'].mean(), inplace=True)
df['sex'].fillna(df['sex'].mode()[0], inplace=True)
df['cp'].fillna(df['cp'].mode()[0], inplace=True)
df['fbs'].fillna(df['fbs'].mode()[0], inplace=True)
df['restecg'].fillna(df['restecg'].mode()[0], inplace=True)
df['exang'].fillna(df['exang'].mode()[0], inplace=True)
df['slope'].fillna(df['slope'].mode()[0], inplace=True)
df['ca'].fillna(df['ca'].mode()[0], inplace=True)
df['thal'].fillna(df['thal'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head(10) #print first 10 rows
```

```python
healthy = df[(df['target'] ==0) ].count()[1]
sick = df[(df['target'] ==1) ].count()[1]
print ("Number of people without heart disease: "+ str(healthy))
print ("Number of people with heart disease: "+ str(sick))
```

```python
X = df.iloc[:,0:13].values 
y = df.iloc[:,13].values # integer-location based indexing for selection by position.
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
X_train.shape, y_train.shape, X_test.shape , y_test.shape
```

```python
corr = df.corr() # calculate the correlation matrix
fig = plt.figure(figsize=(15,12)) # plot the heatmap
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, linewidths=.75, annot=True)
```

```python
results_test = {}
results_train = {}
list_algos=[]
```

```python
def predict_data(algo_name,X_train,y_train,X_test,y_test):
    algo_name.fit(X_train, y_train)
    Y_pred = algo_name.predict(X_test)
    acc_train = round(algo_name.score(X_train, y_train) * 100, 2)
    acc_val = round(algo_name.score(X_test, y_test) * 100, 2)
    
    results_test[str(algo_name)[0:str(algo_name).find('(')]] = acc_val
    results_train[str(algo_name)[0:str(algo_name).find('(')]] = acc_train
    list_algos.append(str(algo_name)[0:str(algo_name).find('(')])
    print("acc train: " + str(acc_train))
    print("acc test: "+ str(acc_val))
```

```python
def conf(algo_name, X_test, y_test):
    y_pred = algo_name.predict(X_test)
    forest_cm = metrics.confusion_matrix(y_pred, y_test)
    sns.heatmap(forest_cm, annot=True, fmt='.2f',xticklabels = ['1', '0'] , yticklabels = ['1', '0'])
    plt.ylabel('True class')
    plt.xlabel('Predicted class')
    plt.title(str(algo_name)[0:str(algo_name).find('(')])
```

```python
# RANDOM FOREST
random_forest = RandomForestClassifier(n_estimators=50, random_state = 0)
predict_data(random_forest,X_train,y_train,X_test,y_test)
conf(random_forest, X_test,y_test)
print(classification_report(y_test, random_forest.predict(X_test)))
```

```python
#KNN

knn = KNeighborsClassifier(n_neighbors = 5, weights='uniform', p=2, metric='euclidean')
predict_data(knn,X_train,y_train,X_test,y_test)
conf(knn, X_test,y_test)
print(classification_report(y_test, knn.predict(X_test)))
```

```python
#Gaussian Naive Bayes

gnb = GaussianNB()
predict_data(gnb,X_train,y_train,X_test,y_test)
conf(gnb, X_test,y_test)
print(classification_report(y_test, gnb.predict(X_test)))
```

```python
df_test =pd.DataFrame(list(results_test.items()),columns=['algo_name','acc_test'])
df_train =pd.DataFrame(list(results_train.items()),columns=['algo_name','acc_train'])
df_results = df_test.join(df_train.set_index('algo_name'), on='algo_name')
df_results.sort_values('acc_test',ascending=False)
```
